# Preparing the data

```python
#import libraries
import numpy as np  # useful for scientific computing in Python
import pandas as pd # primary data structure library
%matplotlib inline 
import matplotlib as mpl  #  library for creating static, animated, and interactive visualizations in Python
import matplotlib.pyplot as plt
print ('Matplotlib version: ', mpl.__version__)
```

```python
#download the data
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data', sep=',', index_col=0)
print ('Data read into a pandas dataframe!')
```

```python
#check the data
df_survey
```

```python
#sort the data by 'very interested'
df_survey.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
df_survey.head()
```

# Bar graph

```python
#add the color list
colors_list = ['#5cb85c','#5bc0de','#d9534f']

#calculate the shares
df_survey_prc = df_survey.div(df_survey.sum(1), axis=0)
df_survey_prc

#create the barplot
ax = df_survey_prc.plot(kind='bar',
                     figsize = (20,8),
                     width = 0.8,
                     color = colors_list,
                     edgecolor=None)
plt.legend(labels = df_survey.columns, fontsize = 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize= 16)

#create the ticks
plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

#Add this loop to add the annotations
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center')
```

# Nerighborhood-people

```python
#read the data
df_crimes = pd.read_csv('https://cocl.us/sanfran_crime_dataset')

print('Dataset downloaded')
```

```python
df_crimes
```

```python
df_crimes['PdDistrict'].value_counts()
```

```python
df_crimes_t = df_crimes.PdDistrict.value_counts()

ten_districts = pd.DataFrame(data=df_crimes_t.values, index=df_crimes_t.index, columns=['Count'])
ten_districts = ten_districts.reindex(["CENTRAL", "NORTHERN", "PARK", "SOUTHERN", "MISSION", "TENDERLOIN", "RICHMOND", "TARAVAL", "INGLESIDE", "BAYVIEW"])

ten_districts = ten_districts.reset_index()
ten_districts.rename({'index': 'Neighborhood'}, axis='columns', inplace=True)

ten_districts
```

```python
sf_geo = r'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/san-francisco.geojson' # geojson file

# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42

# create a world map
sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
#create beautiful map :)
sf_map.choropleth(
    geo_data=sf_geo,
    data=ten_districts,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)

sf_map
```
